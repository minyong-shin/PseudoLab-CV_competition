```
import zipfile
import glob
import os
import shutil
from tqdm.notebook import tqdm
```

```
# 해당 경로에 있는 폴더의 파일을 모두 한 곳으로 모으기 
# exercise
all_path = "/content/drive/My Drive/Dacon-랜드마크 이미지 분류/lanmark_dataset/image_train/"
other = os.listdir(all_path)
```

```
for folder_name in tqdm(other):
    print(f"find {folder_name}")
    # 하위 폴더의 path 설정 
    folder_path = all_path + folder_name
    print(f"folder_path : {folder_path}")
    for file_name in os.listdir(folder_path):
        # 하위 폴더내의 파일을 모두 상위 폴로 이동 
        shutil.move(folder_path + '/' + file_name, all_path + '/' + file_name)
    os.rmdir(folder_path)
    
print("end..")
```

```
# 해당 경로에 있는 폴더의 파일을 모두 한 곳으로 모으기 
# exercise
move_path = '/content/drive/My Drive/Dacon-랜드마크 이미지 분류/lanmark_dataset/test_dataset'
all_path = "/content/drive/My Drive/Dacon-랜드마크 이미지 분류/lanmark_dataset/test_dataset/test"
other = os.listdir(all_path)
```

```
other2 = []
for i in other:
  if 'DS' in i:
    pass
  else:
    other2.append(i)
```

```
for folder_name in tqdm(other2): # 옮길 대상이 되는 파일 경로
    print(f"find {folder_name}")
    # 하위 폴더의 path 설정 
    folder_path = os.path.join(all_path,folder_name)
    print(f"folder_path : {folder_path}")

    for file_name in os.listdir(folder_path):
        shutil.move(folder_path + '/' + file_name, move_path + '/' + file_name)
    os.rmdir(folder_path)
print("end..")
```
